```python
import os
import imageio
import scipy.ndimage as ndi
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
```

```python
path = "/home/yosagaf/devs/medical-biometrics/assignement3/data/"
MRI_images = os.listdir(path)
MRI_images.sort()
```

```python
print(MRI_images)
```

```python
def get_nifti_data(img_loc):
    """return nibabel img and matrix from nibabel loaded nifti
    Key arguments
        img_loc: Pathor string of nifti image
    Returns:
        (numpy array of the data)
    """
    img = nib.load(str(img_loc))
    data = img.get_fdata()
    return data
```

```python
data = [get_nifti_data(path+"/"+s) for s in MRI_images]
```

```python
def view_plot(d):
    fig, ax = plt.subplots(1,3, figsize=(16,10))
    ax[0].imshow(d[20],cmap='gray')
    ax[1].imshow(d[80],cmap='gray')
    ax[2].imshow(d[120],cmap='gray')

    ax[0].axis('off')
    ax[1].axis('off')
    ax[2].axis('off')
```

```python
view_plot(data[6])
```
